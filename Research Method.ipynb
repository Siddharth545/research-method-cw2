```python

data_1=pd.DataFrame()
for i in ["IRL","USA","AUS","JPN","RUS","KOR"]:
    df = pd.DataFrame(columns=['Year', 'Age', 'Female','Male','Total'])
    f = open(f'/Users/siddharthsharma/Desktop/hmd_countries_20230403/{i}/STATS/Exposures_1x1.txt', "r")
    lines = f.readlines()
    f.close()

    # remove /n at the end of each line
    for index, line in enumerate(lines):
              lines[index] = line.split()
    for index,line in enumerate(lines[3:]):
        df.loc[index]=line
    df["country"]= i
    data_1=data_1.append(df)
```
